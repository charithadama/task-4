```python
import cv2
import matplotlib.pyplot as mp
import numpy as np
```

```python
img1=cv2.imread("ironman.jfif")
img2=cv2.imread("spiderman.jpg")
img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
mp.imshow(img1)
```

```python
mp.imshow(img2)
```

```python
img1_array = img1.copy()
img2_array = img2.copy()
```

```python
img1_array_cropped = img1_array[50:600, 600:1000]
img1_array_cropped.shape
```

```python
mp.imshow(img1_array_cropped)
cv2.imwrite("img1.jpg",img1_array_cropped)
```

```python
img2_array_cropped = img2_array[50:600, 600:1000]
img2_array_cropped.shape
```

```python
mp.imshow(img2_array_cropped)
cv2.imwrite("image2.jpg",img2_array_cropped)
```

```python
img1[50:600, 600:1000] = img2_array_cropped
```

```python
mp.imshow(img1)
```

```python
img2[50:600, 600:1000] = img1_array_cropped
```

```python
mp.imshow(img2)
```
