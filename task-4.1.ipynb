```python
import cv2
```

```python
import numpy as np
```

```python
pic=np.zeros((550,500,3))
```

```python
cv2.circle(pic, center=(250,130), radius=100, color=(240,255,100), thickness=-1)
```

```python
cv2.circle(pic,center=(215,120),radius=15,color=(0,0,0),thickness=-1)
cv2.circle(pic,center=(275,120),radius=15,color=(0,0,0),thickness=-1)
```

```python
cv2.circle(pic,center=(250,180),radius=47,color=(0,0,0),thickness=-1)
```

```python
cv2.circle(pic,center=(245,300),radius=150,color=[240,255,100],thickness=-1)
```

```python
cv2.line(pic,(250,390),(150,500),(240,255,100),30)
cv2.line(pic,(250,390),(350,500),(240,255,100),30)
```

```python
cv2.imshow("mypic",pic)
cv2.waitKey()
cv2.destroyAllWindows()
```
